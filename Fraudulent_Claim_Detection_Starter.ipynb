### **4.4 Perform bivariate analysis** <font color="red">[10 Marks]</font>

#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features
import math
def target_likelihood_analysis(X, y, target_value='Y'):
    categorical_cols = X.select_dtypes(include='object').columns
    likelihood_summary = {}

    num_features = len(categorical_cols)
    n_cols = 3  # ⬅️ 3 subplots per row
    n_rows = math.ceil(num_features / n_cols)

    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5 * n_rows))
    axes = axes.flatten()

    for i, col in enumerate(categorical_cols):
        df_temp = pd.concat([X[col], y], axis=1)
        stats = (
            df_temp.groupby(col)[y.name]
            .agg(
                count='count',
                target_count=lambda x: (x == target_value).sum()
            )
            .assign(likelihood=lambda df: df['target_count'] / df['count'])
            .sort_values('likelihood', ascending=False)
        )
        likelihood_summary[col] = stats

        sns.barplot(x=stats.index, y=stats['likelihood'], ax=axes[i], palette='crest')
        axes[i].set_title(f"Target Likelihood by '{col}'")
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Likelihood')
        axes[i].tick_params(axis='x', rotation=45)

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()

    return likelihood_summary
likelihoods = target_likelihood_analysis(X_train, y_train)
```

```python
variation_summary = {}

for feature, stats in likelihoods.items():
    variation = stats['likelihood'].std()
    variation_summary[feature] = variation

# Sort by lowest variation
low_variation = sorted(variation_summary.items(), key=lambda x: x[1])

# Sort by highest variation
high_variation = sorted(variation_summary.items(), key=lambda x: x[1], reverse=True)

# Create a DataFrame for easier plotting
df_variation = pd.DataFrame(high_variation, columns=['feature', 'std'])

# Compute cumulative variance
df_variation['cumulative_variance'] = df_variation['std'].cumsum()

# Normalize if you want the cumulative percentage (optional)
df_variation['cumulative_percentage'] = 100 * df_variation['cumulative_variance'] / df_variation['std'].sum()

# Plot
plt.figure(figsize=(12, 6))
plt.plot(df_variation['feature'], df_variation['cumulative_percentage'], marker='o', label='Cumulative % Variance')
plt.xticks(rotation=90)
plt.title("Cumulative Variance of Categorical Features (Descending Order)")
plt.xlabel("Categorical Feature (sorted by std)")
plt.axhline(y=80, color='red', linestyle='--', label='80% Threshold')  # Optional: Pareto threshold
plt.ylabel("Cumulative % of Variance")
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()

# Display Top 5
print("\n🔥 Top 5 Categorical Features with Highest Variation in 'Y' Likelihood:")
for feature, std_val in high_variation[:5]:
    print(f"{feature}: likelihood std = {std_val:.4f}")
    
print("\n🔍 Categorical features with the **least** variation in 'Y' likelihood:")
for feature, std_val in low_variation[:5]:  # Show top 5 weakest
    print(f"{feature}: likelihood std = {std_val:.4f}")
```

### **6.2 Feature Creation** <font color="red">[4 marks]</font>

Create new features from existing ones to enhance the model's ability to capture patterns in the data. This may involve deriving features from date/time columns, combining features, or creating interaction terms.

```python
# Create new features based on your understanding for both training and validation data
def create_features(df):
    df = df.copy()

    # --- Convert dates to datetime
    df['policy_bind_date'] = pd.to_datetime(df['policy_bind_date'], errors='coerce')
    df['incident_date'] = pd.to_datetime(df['incident_date'], errors='coerce')

    # --- Extract date parts
    df['policy_bind_year'] = df['policy_bind_date'].dt.year
    df['policy_bind_month'] = df['policy_bind_date'].dt.month
    df['policy_bind_dayofweek'] = df['policy_bind_date'].dt.dayofweek
    df['incident_year'] = df['incident_date'].dt.year
    df['incident_month'] = df['incident_date'].dt.month
    df['incident_dayofweek'] = df['incident_date'].dt.dayofweek

    # --- Policy age in years (from bind date to incident date)
    df['policy_age_years'] = (df['incident_date'] - df['policy_bind_date']).dt.days // 365

    # --- Extract and combine policy_csl
    df[['csl_per_person', 'csl_per_accident']] = df['policy_csl'].str.split('/', expand=True).astype(float)
    df['csl_total'] = df['csl_per_person'] + df['csl_per_accident']
    df['auto_full_name'] = df['auto_make'].astype(str) + '-' + df['auto_model'].astype(str)
    
    # --- Ratios and interactions
    df['claim_per_month'] = df['total_claim_amount'] / df['months_as_customer'].replace(0, 1)
    df['vehicle_claim_ratio'] = df['vehicle_claim'] / df['total_claim_amount'].replace(0, 1)
    df['injury_claim_ratio'] = df['injury_claim'] / df['total_claim_amount'].replace(0, 1)
    df['property_claim_ratio'] = df['property_claim'] / df['total_claim_amount'].replace(0, 1)

    # --- Financial features
    df['net_capital'] = df['capital-gains'] - df['capital-loss']
    df['high_premium'] = (df['policy_annual_premium'] > df['policy_annual_premium'].quantile(0.9)).astype(int)
    df['high_claim'] = (df['total_claim_amount'] > df['total_claim_amount'].quantile(0.9)).astype(int)

    # --- Combined risk feature: policy_deductable vs umbrella_limit
    df['deductible_to_umbrella_ratio'] = df['policy_deductable'] / (df['umbrella_limit'] + 1)
    return df
X_train_resampled = create_features(X_train_resampled)
X_val = create_features(X_val)
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information
def combine_low_frequency_categories(df, column, threshold=0.02):
    df = df.copy()
    value_counts = df[column].value_counts(normalize=True)
    rare_categories = value_counts[value_counts < threshold].index
    df[column] = df[column].apply(lambda x: 'Other' if x in rare_categories else x)
    return df
```
